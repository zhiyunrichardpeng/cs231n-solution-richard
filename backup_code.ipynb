```python
from google.colab import drive, files
import os
import shutil


# 1. Mount Google Drive
drive.mount('/content/drive')

import datetime

# --- Corrected Python Code ---
# 1. Get the current date and time
now = datetime.datetime.now()

# 2. Format the time object into the desired string (YYYYMMDD_HHMMSS)
timestamp = now.strftime('%Y%m%d_%H%M%S')

# 3. Create the automated, unique file name
NEW_NOTEBOOK_NAME = f"Backup_after_from_instruction_{timestamp}.ipynb"

# 2. Define the save path
DRIVE_FOLDER = "/content/drive/My Drive/cs231n/assignments/assignment1/backup/"
# NEW_NOTEBOOK_NAME = "Backup_Notebook_$(date '+%Y%m%d_%H%M%S').ipynb" # Naming with a timestamp for easy tracking

# 3. Find the current notebook's path (requires Colab's internal structure)
try:
    # This is the most reliable way to get the *temporary* file path of the running notebook
    notebook_path = "/content/drive/My Drive/cs231n/assignments/assignment1/knn.ipynb"
    print(f"Source path identified: {notebook_path}")

    # 4. Prepare the target path
    os.makedirs(DRIVE_FOLDER, exist_ok=True) # Ensure the directory exists
    target_path = os.path.join(DRIVE_FOLDER, NEW_NOTEBOOK_NAME)

    # 5. Copy the file
    shutil.copy(notebook_path, target_path)

    print("-" * 40)
    print(f"✅ Success! Notebook saved to Google Drive at:")
    print(f"**{target_path}**")

except Exception as e:
    print(f"❌ An error occurred during the save process: {e}")
    print("Please ensure your notebook has been saved once via the File > Save a copy menu before running this code.")
```
